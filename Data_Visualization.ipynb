5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.



Ans-  To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap in Seaborn, we can use the following code. This code creates a dataset with multiple features that have a defined correlation and then visualizes the correlation matrix using Seaborn's heatmap function.

Here's how to implement this:

import numpy as np  
import pandas as pd  
import seaborn as sns  
import matplotlib.pyplot as plt  

# Set a random seed for reproducibility  
np.random.seed(0)  

# Generate synthetic data with correlated features  
n_samples = 100  
feature_1 = np.random.rand(n_samples)  # First feature  
feature_2 = feature_1 * 0.5 + np.random.rand(n_samples) * 0.5  # Second feature correlated with the first  
feature_3 = feature_1 * 0.3 + feature_2 * 0.7 + np.random.rand(n_samples) * 0.1  # Third feature correlated with the first two  
feature_4 = np.random.rand(n_samples)  # Fourth feature uncorrelated  

# Create a DataFrame  
data = pd.DataFrame({  
    'Feature 1': feature_1,  
    'Feature 2': feature_2,  
    'Feature 3': feature_3,  
    'Feature 4': feature_4  
})  

# Calculate the correlation matrix  
correlation_matrix = data.corr()  

# Create a heatmap to visualize the correlation matrix  
plt.figure(figsize=(8, 6))  
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)  

# Title  
plt.title('Correlation Matrix Heatmap')  

# Display the plot  
plt.show()  

Explanation:

Imports: The necessary libraries (NumPy, Pandas, Seaborn, and Matplotlib) are imported.

Random Seed: A random seed is set for reproducibility.

Synthetic Data Generation:
Four features are generated. Feature 1 is random, Feature 2 is correlated with Feature 1, and Feature 3 is a combination of Feature 1 and Feature 2 plus some noise. Feature 4 is independent and uncorrelated.

DataFrame: A Pandas DataFrame is created to hold the features.

Correlation Matrix: The correlation matrix is calculated using the .corr() method.

Heatmap: The correlation matrix is visualized using sns.heatmap(), with:
annot=True to display the correlation values on the heatmap.
cmap to select the color palette (in this case, 'coolwarm').
fmt='.2f' to format the annotations to two decimal places.
linewidths to add separation between squares in the heatmap for better clarity.

Title: A title is added for context.

Display: Finally, the heatmap is displayed using plt.show().

import numpy as np

# Create a DataFrame (optional, as Plotly can work directly with numpy arrays)  
data = {

    'Y': y.flatten(),  
    'Z': z.flatten()  
}

                  scene=dict(  
                      xaxis_title='X Axis',  
                      yaxis_title='Y Axis',  
                      zaxis_title='Z Axis'  
                  ))  

# Show the figure

import numpy as np  
from bokeh.plotting import figure, show  
from bokeh.io import output_file  
from bokeh.transform import linear_cmap  
from bokeh.models import ColorBar  
from bokeh.tile_sources import get_provider, Vendor  

# Generate random data for the heatmap  
data_heatmap = np.random.rand(10, 10)  
x = np.linspace(0, 1, 10)  
y = np.linspace(0, 1, 10)  
xx, yy = np.meshgrid(x, y)  

# Prepare data for Bokeh  
x_coords = xx.flatten()  
y_coords = yy.flatten()  
values = data_heatmap.flatten()  

# Create a color mapper  
mapper = linear_cmap(field_name='value', palette="Viridis256", low=min(values), high=max(values))  

# Prepare the data source  
source = ColumnDataSource(data=dict(x=x_coords, y=y_coords, value=values))  

# Output to a static HTML file  
output_file("heatmap.html")  

# Create a Bokeh figure  
p = figure(title="Heatmap of Random Data",   
           tools="reset",  
           x_range=(0, 1),   
           y_range=(0, 1),  
           x_axis_label='X Axis',   
           y_axis_label='Y Axis')  

# Add squares to the figure for the heatmap  
p.rect(x='x', y='y', width=0.1, height=0.1, source=source,   
        line_color=None, fill_color=mapper)  

# Add a color bar  
color_bar = ColorBar(color_mapper=mapper['field'], width=8, location=(0,0))  
p.add_layout(color_bar, 'right')  

# Show the result  
show(p)  


Explanation:

Data Generation: The code generates a 10x10 matrix data_heatmap filled with random values.

Meshgrid Creation: np.meshgrid creates two-dimensional arrays for the x and y coordinates.

Data Preparation: The x and y coordinates along with the corresponding values from data_heatmap are flattened to prepare for plotting.

Data Source: A ColumnDataSource is created from the flattened x, y, and value arrays.

Output: The output_file function specifies the name of the HTML file for the output.

Bokeh Figure: A figure is created with a specified title, tools, and axis labels.

Rectangles for Heatmap: The rect method adds colored rectangles (0.1 by 0.1) to represent the data values in a grid format.

Color Bar: A ColorBar is added to provide a reference for the color values.

Displaying the Plot: The show function is called to render the heatmap in our web browser.
